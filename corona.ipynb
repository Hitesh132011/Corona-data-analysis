```python
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], axis=1, inplace=True)
```

```python
covid_df.head()
```

```python
covid_df['Date']=pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
#Active cases
covid_df['Active_Cases']=covid_df['Confirmed']-covid_df['Cured']-covid_df['Deaths']
```

```python
covid_df.head()
```

```python
statewise=pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured"],index="State/UnionTerritory",aggfunc=max)
```

```python
statewise['Recovery Rate']=statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise['Mortality Rate']=statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise=statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
#Top 10 active cases
top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by=['Active_Cases'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 states with most active cases in India",size=25)
```

```python
print(top_10_active_cases.columns)
```

```python
ax = sns.barplot(data=top_10_active_cases.iloc[:10], y="Active_Cases", x="State/UnionTerritory", linewidth=2, edgecolor='red')
```

```python
#Top 10 active cases
top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by=['Active_Cases'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 states with most active cases in India",size=25)
ax = sns.barplot(data=top_10_active_cases.iloc[:10], y="Active_Cases", x="State/UnionTerritory", linewidth=2, edgecolor='red')
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
top_10_deaths = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(18,9))
plt.title("Top 10 states with most Death in India",size=25)
ax = sns.barplot(data=top_10_deaths.iloc[:10], y="Deaths", x="State/UnionTerritory", linewidth=2, edgecolor='red')
plt.xlabel("States")
plt.ylabel("Deaths")
plt.show()
```

```python
fig = plt.figure(figsize=(12, 6))
states = ['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh']
ax = sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(states)],x='Date',y='Active_Cases',hue='State/UnionTerritory')
ax.set_title("Top 5 Affected States in India", size=16)
plt.show()
```

```python
vaccine=vaccine_df[vaccine_df.State!='India']
```

```python
vaccine.head()
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
vaccine.head()
```

```python
#Most Vaccinated State
max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values('Total',ascending=False)
max_vac
```

```python
fig = plt.figure(figsize=(10, 5))
plt.title("Top Vaccinated States in India", size=20)
ax = sns.barplot(data=max_vac.iloc[:5], y="Total", x=max_vac.index[:5], linewidth=2, edgecolor='red')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
plt.title("Least Vaccinated States in India", size=20)
ax = sns.barplot(data=max_vac.tail(5), y="Total", x=max_vac.tail(5).index, linewidth=2, edgecolor='red')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
